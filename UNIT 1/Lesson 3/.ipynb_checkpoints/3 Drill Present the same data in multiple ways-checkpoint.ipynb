```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
crash = sns.load_dataset("car_crashes")
```

```python
crash
```

```python
# plot variable "total" in four different ways

# 1. plot
plt.plot(crash['total'])
plt.show()
```

```python
# 2. scatter
plt.scatter(x = crash['speeding'], y = crash['total'])
plt.title('Total crashes for speeding')
plt.xlabel('Speeding')
plt.ylabel('Total Crashes')
plt.show()
```

```python
# 3. bargraph
sns.set(style="darkgrid")

g = sns.factorplot(x="abbrev", y="total", data=crash,
                   size=50, kind="bar", palette="pastel", ci=95)
g.despine(left=True)
g.set_ylabels("Crash")
g.set_xlabels("State")
plt.title('Barplot: Total crashes by State')
plt.show()
```

```python
# 4. PointPlot
sns.set(style="whitegrid")

g = sns.factorplot(x="abbrev", y="total", data=crash,
                   size=20, kind="point", palette="pastel",ci=95,dodge=True,join=False)
g.despine(left=True)
g.set_ylabels("total")
g.set_xlabels("state")
plt.title('Pointplot: total crash by state')
plt.show()
```
