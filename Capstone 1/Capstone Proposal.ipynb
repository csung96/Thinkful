```python
# Data Cleaning Necessary
# there is a lot of NaN values, so it seems like there will need to be a lot of data cleaning and organizing
```

```python
# Make sure Age is integer, not string
df['Age'].apply(pd.to_numeric, errors='ignore')
```

```python
# Demographis: AGE
plt.figure(figsize=(30,10))
sns.countplot(df["Age"])
plt.show()
```

```python
# Demographics: Gender
plt.xticks(rotation='vertical')
sns.countplot(df['Gender'])

# Need to clean the data to either Female, Male, or Other
```

```python
# clean gender to "Female", "Male", "Others"
df['Gender'].str.upper()
df['Gender'] = df['Gender'].map({'FEMALE': 'Female'})
df['Gender'].value_counts()
```

```python
# Demographics: Country
sns.countplot(df['Country'])
plt.xticks(rotation='vertical')

# most people are from US, Canada, and UK
# therefore this sample is not a good representation for the whole population
# most conclusions should be concerning US, UK and canada
```

```python
# dataset not representative of global population since from the graph, 
# we can see that most of the sample is from US, UK and Canada. 
# Therefore, any conclusions/results from the analysis can only really be 
# implied/applied to the top three countries since that what our sample represents

df.groupby(df['Country']).count()
```

```python
# Mental Health condition--> is there a need for mental health resources in workplace?
#### Family history: do you have family history of mental illenss?
#### Treatment: have you sought treatment for mental condition?
#### work interference: if you have mental health condition, do you feel it interferes with work?
```

```python
# use the top three countries
# US = 751
# UK = 185
# Canada = 72
# Total in top 3 countries = 1008
countries = pd.concat([df.loc[df['Country']=='Canada'], df.loc[df['Country']=='United States'], df.loc[df['Country']=='United Kingdom']]).reset_index(drop=True)
print('There consists {} people from the top 3 countries out of the {} people surveyed'.format(countries.shape[0], df.shape[0]))
```

```python
# Does the employer recognize importance of mental health
# Does employer offer resources to employees for mental health issues
```

```python
sns.countplot(x=countries['work_interfere'], order=['Never', 'Rarely', 'Sometimes', 'Often'])
plt.xlabel('If you have a mental health condition, how often do you feel that it interferes with your work?')
```

```python
sns.countplot(countries['treatment'])
plt.xlabel('Have you sought treatment for a mental health condition?')
```

```python
countries.groupby(countries['treatment']).count()
```

```python
# 540 individuals (53.57%) have sought treatement 
```
