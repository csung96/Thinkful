Dataset = sf airbnb listings from 7/10/2017

Variables: 
1. room_id: A unique number identifying an Airbnb listing.
2. host_id: A unique number identifying an Airbnb host. 
3. room_type: One of “Entire home/apt”, “Private room”, or “Shared room”
4. borough: A subregion of the city or search area for which the survey is carried out. The borough is taken from a shapefile of the city that is obtained independently of the Airbnb web site. For some cities, there is no borough information; for others the borough may be a number. If you have better shapefiles for a city of interest, please send them to me.
5. neighborhood: As with borough: a subregion of the city or search area for which the survey is carried out. For cities that have both, a neighbourhood is smaller than a borough. For some cities there is no neighbourhood information.
6. reviews: The number of reviews that a listing has received. Airbnb has said that 70% of visits end up with a review, so the number of reviews can be used to estimate the number of visits. Note that such an estimate will not be reliable for an individual listing (especially as reviews occasionally vanish from the site), but over a city as a whole it should be a useful metric of traffic.
7. overall_satisfaction: The average rating (out of five) that the listing has received from those visitors who left a review.
8. accommodates: The number of guests a listing can accommodate.
9. bedrooms: The number of bedrooms a listing offers.
10. price: The price (in $US) for a night stay. In early surveys, there may be some values that were recorded by month.
11. minstay: The minimum stay for a visit, as posted by the host.
12. latitude and longitude: The latitude and longitude of the listing as posted on the Airbnb site: this may be off by a few hundred metres. I do not have a way to track individual listing locations with
13. last_modified: the date and time that the values were read from the Airbnb web site.

Exploring the data: Bivariate

1. Two continuous variables: reviews and overall_satisfaction
        * Scatterplot matrix: to show all bivariate relationships in 
          one plot made of subplots. 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats as stats
```

```python
# set default plot to be clearer and easier to read
sns.set_style("white")

# subset of data to make plots clearer (only use cont variables)
df_cont = df.loc[ : ,['neighborhood', 'reviews', 'overall_satisfaction', 'accommodates',
                      'bedrooms','price']]
df_cont
```

```python
# To see all the points--> create df of rand noise (jitter)
# and add it only when plotting

# Make a copy of the data to add jitter to and plot.
df_jittered = df_cont.loc[:, 'reviews':'price'].dropna()

# Make the random noise.
jitter = pd.DataFrame(
    np.random.uniform(-.3, .3, size=(df_jittered.shape)),
    columns=df_jittered.columns
)

# Combine the data and the noise.
df_jittered = df_jittered.add(jitter)
```

```python
# Declare that we are making scatterplot matrix.
g = sns.PairGrid(df_jittered.dropna(), diag_sharey=False)
# Scatterplot
g.map_upper(plt.scatter, alpha=.5)
# Fit line summarizing the linear rel of the two variables.
g.map_lower(sns.regplot, scatter_kws=dict(alpha=0))
# Give info about the univariate distributions of the variables.
g.map_diag(sns.kdeplot, lw=3)
plt.show()
```

```python
# Make the correlation matrix.
corrmat = df_cont.corr()
print(corrmat)

# Set up the matplotlib figure.
f, ax = plt.subplots(figsize=(12, 9))

# Draw the heatmap using seaborn.
sns.heatmap(corrmat, vmax=.8, square=True)
plt.show()
# Heatmaps = great way to get visual read on corr matrix

```

For continuous and categorical pair: 

* Descriptive stats (est of central tendency and variance) for continuous variable at each level of categorical variable

* T-tests and ANOVA

```python
# Plot all the variables with boxplots for each neighborhood

# Restructure the data so we can use FacetGrid 
# rather than making a boxplot for each variable separately.
df_long = df_cont
df_long = pd.melt(df_long, id_vars=['neighborhood'])


g = sns.FacetGrid(df_long, col="variable", size= 10, aspect=.5)
g = g.map(sns.boxplot, "neighborhood", "value")
plt.show()
```

```python
# Descriptive statistics by group.
print(df_cont.groupby('neighborhood').describe())
# Test whether group differences are significant.
for col in df_cont.loc[:,'reviews':'price'].columns:
    print(col)
    print(stats.ttest_ind(
        df_cont[df_cont['neighborhood'] == 'Bayview'][col].dropna(),
        df_cont[df_cont['neighborhood'] == 'Bernal Heights'][col].dropna(),
    ))
```
