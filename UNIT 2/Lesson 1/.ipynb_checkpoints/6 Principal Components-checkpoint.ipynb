PRINCIPAL COMPONENTS ANALYSIS (PCA)
complexcity reduction technique that tries to reduce a set of varibles down to a smaller set of comp that rep most of info in the variables
workd by identifying sets of var that share variance and creating a comp to rep that variance
losing variance in exchange for a smaller set of features can be worthwhile
some models (like reg) assume that features will be uncorr with each other and high levels of inter-feature corr create unstable solutions
solutions with fewer features are easier to understand and are more computationally efficient
solutions with fewer features are also less vulnerable to overfitting


THINGS GET MESSY
PCA works best for norm dist data and assumes rel among var are linear
also works best when variables involved range from weakly corr to moderately strongly corr
correlations of 0 or corr greater than .7/.8 create prob
corr near - mean there is no shared info
corr that are too high can result in comp that are unstable and change a lot in response to very small changes in data

PCA: ROTATION IN SPACE
picture dataset of n variables as an n-dimentsional space
this is generalization of 2-dim scatterplot we re familiar with (rep two variables in a two dim space) 
in PCA, variables are first standardized to have a mean of 0 and std dev of 1
now all variables go through the origin point (where the value of all axes is 0) and share same variance
next, axes of n-dim space are rotated to minimize the distance b/w datapoints and the axes
during rotation some axes become shorter, indicating that variacne along that axis is small and the axis contains little info


```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
from matplotlib.mlab import PCA as mlabPCA
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA 
```

```python
# Make the scatterplot matrix.

# Setting the default plot aesthetics to be prettier.
sns.set_style("white")

df = pd.read_csv('https://raw.githubusercontent.com/Thinkful-Ed/data-201-resources/master/ESS_practice_data/ESSdata_Thinkful.csv')

# Take a subset of the data for PCA. This limits to Swiss and Czech data from 2012
# and keeps only specific columns.
df_pca = df.loc[
    ((df['cntry'] == 'CZ') | (df['cntry'] == 'CH')) & (df['year'] == 6),
    ['tvtot','ppltrst','pplfair','pplhlp']
].dropna()

t = sns.regplot(
    'ppltrst',
    'pplfair',
    df_pca,
    x_jitter=.49,
    y_jitter=.49,
    fit_reg=False
)
t.set(xlim=(-1, 11), ylim=(-1, 11))
t.axhline(0, color='k', linestyle='-', linewidth=2)
t.axvline(0, color='k', linestyle='-', linewidth=2)
t.axes.set_title('Raw data')
plt.show()

# Standardizing variables by subtracting the mean and dividing by the standard
# deviation. Now both variables are on the same scale.
df_pca['ppltrst_z'] = (df_pca['ppltrst'] - df_pca['ppltrst'].mean()) / df_pca['ppltrst'].std()
df_pca['pplfair_z'] = (df_pca['pplfair'] - df_pca['pplfair'].mean()) / df_pca['pplfair'].std()

t = sns.regplot(
    'ppltrst_z',
    'pplfair_z',
    df_pca,
    x_jitter=.49,
    y_jitter=.49,
    fit_reg=False
)
t.axhline(0, color='k', linestyle='-', linewidth=2)
t.axvline(0, color='k', linestyle='-', linewidth=2)
t.axes.set_title('Standardized data')
plt.show()
```

in plot above, both axes describe equal variance they both run from about -3 to 2.5
if we were to drop one of the axes and describe the data using only the info from the other axis, we would lose roughly 50% of our info
but if we fit line through origint hat minimizes distance b/w line and each point and then rotat data, our axes and info will change

```python
# Rotating the data so that the axes roughly match to vector above.
df_pca['ppltrst_z_rot'] = math.cos(40) * df_pca['ppltrst_z'] - math.sin(40) * df_pca['pplfair_z']
df_pca['pplfair_z_rot'] = math.sin(40) * df_pca['ppltrst_z'] + math.cos(40) * df_pca['pplfair_z']

t = sns.regplot(
    'ppltrst_z_rot',
    'pplfair_z_rot',
    df_pca,
    x_jitter=.49,
    y_jitter=.49,
    fit_reg=False
)
t.axhline(0, color='k', linestyle='-', linewidth=2)
t.axvline(0, color='k', linestyle='-', linewidth=2)
t.axes.set_title('Rotated standardized data')
plt.show()
```

in image aboce, horizontal vector is longer than vertical vecor (spanning rom -3.5 to 4 rather than -3 to 3.5)
horizontal axis contains more variance and has a higher info value than vertical vector
if we treated axes as principal components and discarded vertical vector, each data point would be described solely by its position on the horizontal axis. 
info on each datapoint's vertical posistion woud be lost
because of the rotation if we drop the vertical axis now, we lose_less_ than 50% of the info

THE MATH BEHIND PCA:

The rotation pictured above is based on calculating the covariance matrix of the data and applying a linear trans (rotation) and deriving from that the eigenvalues and eigen vectors that express amount of varianc in the data explained by the new axes. 

COVARIANCE MATRIX:

Corr matrix = covariance matrix where covariances have been divided by the variances this standardizes the covariances so that they are all on the same scale (-1,1) and can be compared covariance matrices (like corr mat) contain info about the amount of variance shared b/w pairs of variables

variance of x is sum of the sq diff b/w each value in x (xi) and the mean of x ( x bar) divided by sample size (n)
cov b/w two variables x and y is the sum of product of the differences of each variable value and its mean, divide by the sample size 

```python
# covariance matrix for ESS data: 

# Take a subset of the data for PCA and drop missing values because PCA cannot
# handle them. We could also impute, but missingness is quite low so dropping
# missing rows is unlikely to create bias.
df_pca = df.loc[
    ((df['cntry'] == 'CZ') | (df['cntry'] == 'CH')) & (df['year']==6),
    ['tvtot','ppltrst','pplfair','pplhlp']
].dropna()

# Normalize the data so that all variables have a mean of 0 and standard deviation
# of 1.
X = StandardScaler().fit_transform(df_pca)

# The NumPy covariance function assumes that variables are represented by rows,
# not columns, so we transpose X.
Xt = X.T
Cx = np.cov(Xt)
print('Covariance Matrix:\n', Cx)
```

DATA AS A MATRIX

PCA deals with ammy variables at once 
and represents those variables in formulas as a matrix with m rows and n columns
where each row rep a measuremant instance (say data from one participant over many variables)
and each column represents a variable
The mathmatical convention is that matrices are referred to in capital letters
ex: a dataset with 3 var and 4 part would be expressed as 4 x 3 matrix X 
(notice that rows come first, then columns)

matrices behave differently than individual variables in many ways. 
in particular, multipllying two matrices together doesnt behave at all like multiplying 2 numbers

because of way matrix mult works, we can express covariance matrix for X as the product of X and transpose of X (X^T)
a matrix is transposed by flipping it on its diagonal so that row and comlumn indices are switched
    1 2 3
    2 3 6
    2 2 9
    4 5 1 
    
    IS NOW
    
    1 2 2 4
    2 3 2 5
    3 6 9 1
    
the covariance matrix for X by logic of matrix mult is an m by m matrix where diag terms rep the variance of each var and the off diag are covariance

ESS data: 
created by multiplying a matrix of data with dimensions 1406 x 4 (1406 rows and 4 variables) 
by the transpose of that matrix, with dimensions 4 x 1406. 
The covariance matrix, as a result, has dimensions 4 x 4.



BACK TO PCA 

* In PCA, goal = transform covariance matrix so that all info is in diagnoals (varaiances) and none is in the off diag (covariances)
* do this by finding a matrix P where when we multiply P and X we get a matrix Y with covariance matrix Cy with only 0s on the off diag

EIGENVECTORS
* Directional aspect of a component
* It is the red line in the graph earlier
* During PCA, they are chosen to be orthogonal (have corr of 0 with onw another0 
* First vector is found that minimizes distance b/w that vector and datapoints- this vector is 1st comp
* 2nd vector is found that also minimizes distance b/w vector and datapoints (vector must be perp to first one)

EIGENVALUES
* Represent the length of the Eigenvectors 
* Each eigenvector has an eigenvalue. 
* Length of the eigenvector encodes the prop of total variance explained by a component. 
* Total variance = to the number of variables in the PCA. 
* Thus, an Eigenvalue of 1 means that the component explains the same amount of variance as one variable. 
* Eigenvalue > 1 is desirable, since a component with an eigenvalue of 1 adds no value beyond the information contained in any individual variable, * Eigenvalue < 1 is less efficient at conveying info than a variable by itself. 
* Eigenvalue of 2 means that the component contains an amount of information equal to that of two variables. 

Prop of the variance that each eigenvector rep can be calc by dividing the eigenvalue corresponding to that eigenvector by the sum of all eigenvalues.

```python
# Eigenvectors, eigenvalues, and variance explained for the ESS data:

# Calculating eigenvalues and eigenvectors.
eig_val_cov, eig_vec_cov = np.linalg.eig(Cx)

# Inspecting the eigenvalues and eigenvectors.
for i in range(len(eig_val_cov)):
    eigvec_cov = eig_vec_cov[:, i].reshape(1, 4).T
    print('Eigenvector {}: \n{}'.format(i + 1, eigvec_cov))
    print('Eigenvalue {}: {}'.format(i + 1, eig_val_cov[i]))
    print(40 * '-')

print(
    'The percentage of total variance in the dataset explained by each',
    'component calculated by hand.\n',
    eig_val_cov / sum(eig_val_cov)
)
```

COMPONENTS
* components = eigenvectors that have been divided by the sq rt of eigenvalues
* this standardizes comp so that amount o variance in each var explained by a comp can be compared across components
* components will replace the factored variables as all the components of a given solution

SUMMARY
* PCA = series of linear transformations applied to df to yield a smaller number of columns that explain a large prop of the total variance contained in the df
* works by transforming data X into a covariance matrix Cx 
* then a set of eigenvectors P is calc that when combined with Cx through mult (PCxP^T), create new cov matrix Cy with 0's on off diag elements
